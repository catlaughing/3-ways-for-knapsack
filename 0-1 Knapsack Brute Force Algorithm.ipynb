```python
from itertools import combinations
from time import time

def brute_force(n, W_max, items):
    best_value = None
    best_combination = []

    for i in range(n):
        print(f'Kombinasi dari {i+1} item')
        for kombinasi in combinations(items, i+1):
            weight = sum([item_w[0] for item_w in kombinasi])
            value = sum([item_v[1] for item_v in kombinasi])
            print(f'     Weight = {weight} | Value = {value} | Kombinasi = {kombinasi}')
            if((best_value is None or best_value < value) and weight <= W_max):
                best_value = value
                best_combination = [0] * n
                for item in kombinasi:
                    best_combination[items.index(item)] = 1
                print(f'          Best combination = {best_combination}')

    return best_value, best_combination
```
